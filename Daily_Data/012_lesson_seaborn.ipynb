## Distribution Plots

#### Rugplot, Displot, Histplot, Kdeplot, Dimensional kdeplot

```python
sns.rugplot(x= "salary",data=df);
```

```python
plt.figure(figsize=(12,8))
sns.rugplot(x= "salary",data=df, height=0.8, color="r", alpha=0.6);
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = "salary", y= "sales", data= df);
sns.rugplot(x= "salary", y= "sales", data=df, height=0.1, color="r", alpha=0.6);
```

```python
sns.displot(data=df, x= "salary", kde=True, bins=10);
```

```python
sns.histplot(data=df, x= "salary", kde=True, bins=10);
```

```python
plt.figure(figsize=(12,8))
sns.histplot(data=df, x= "salary", y="sales", bins=50, color="r");
sns.kdeplot(data=df, x= "salary", y="sales", color="g", alpha=0.6)
```

```python
sns.kdeplot(data=df, x= "salary", bw_adjust=0.2);
```

```python
sns.displot(x= df["salary"],  bins=10, color="r", edgecolor="black", ls="--", lw=2);
```

```python
sns.displot(x= df["salary"],  bins=10, color="r", edgecolor="black", ls="--", lw=2, 
            rug=True, rug_kws={"color":"blue"});
```

```python
sns.displot(x= df["salary"],  bins=10, color="r", kind="hist", edgecolor="black", ls="--", lw=2, 
            rug=True, rug_kws={"color":"blue"});
```

```python
sns.displot(x= df["salary"], y= df["sales"],  bins=10, color="r", edgecolor="black", ls="--", lw=2, 
            rug=True, legend=True);
```

```python
sns.histplot(data=df, x= "salary", bins=10, stat="frequency")
```

```python
sns.histplot(data=df, x= "salary", bins=10, stat="probability", fill =False, element="step")
```

```python
sns.histplot(data=df, x= "salary", bins=10, stat="probability", fill =False, element="step", cumulative=True)
#Estimated Cumulative Distribution Function
```

```python
plt.figure(figsize=(12,8))
sns.histplot(data=df, x= "salary", hue="division", multiple="stack")
```

```python
plt.figure(figsize=(12,8))
sns.histplot(data=df, x= "salary",   hue="training level", multiple="dodge", shrink=0.5);
```

```python
plt.figure(figsize=(12,8))
sns.histplot(data=df[(df["salary"]>90000)&(df["salary"]<100000)], x="salary", 
             hue="division",multiple="dodge",shrink=0.5,bins=5);
```

```python
plt.figure(figsize=(12,8))
ax = sns.histplot(data=df[(df["salary"]>90000)&(df["salary"]<100000)], x="work experience", 
             hue="division",multiple="dodge",shrink=0.5,bins=5);
sns.move_legend(ax, "lower center", bbox_to_anchor=(0.8,0.5,1,1), ncol=1, title="Division", frameon=False)
for i in range(5):
    ax.bar_label(ax.containers[i], label_type="edge", padding=0.1);
```
