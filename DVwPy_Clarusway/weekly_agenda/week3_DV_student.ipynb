```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Set darkgrid style from seaborn

```python
sns.set_style("darkgrid")
```

Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

# 2. Make a bar plot

```python
fig, ax = plt.subplots()
p = np.arange(len(index))
ax.bar(p-0.2, income, width = 0.4, label = "income")
ax.bar(p+0.2, expense, width = 0.4, label = "expense" )
ax.set_xticks(p)
ax.set_xticklabels(index)
plt.legend()
```

```python
ypos = np.arange(len(index))
plt.xticks(ypos, index)
plt.bar(ypos-0.2, income, width=0.4, label = "income")
plt.bar(ypos+0.2, expense, width=0.4, label = "expense")
plt.legend()
```

# 3. Stack the bars

```python
fig, ax = plt.subplots()
p = np.arange(len(index))

ax.bar(p, income,  label = "income")
ax.bar(p,  expense,  label = "expense" ,bottom= income)
ax.set_xticks(p)
ax.set_xticklabels(index)
plt.legend()
```

# 3. Rotate the labels and set figsize

```python
fig, ax = plt.subplots(figsize=(9,6))
p = np.arange(len(index))

ax.bar(p, income,  label = "income", width = 0.5)
ax.bar(p,  expense,  label = "expense" , width = 0.5, bottom= income)
ax.set_xticks(p)
ax.set_xticklabels(index)
plt.legend()
```

# 4. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"
# 5. Unstuck the bars, annotate the hight of the bars on top of them

```python
fig, ax = plt.subplots(figsize=(9,6))
p = np.arange(len(index))
labels = ["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
ax.bar(p-0.1, income,  label = "income", width = 0.2)
ax.bar(p+0.1,  expense,  label = "expense" , width = 0.2)
ax.set_xticks(p)
ax.set_xticklabels(labels)
plt.legend()
```
