```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv") 
```

```python
sns.jointplot(x="math score", y="reading score", data=df);
```

```python
sns.jointplot(x="math score", y="reading score", kind="hex", data=df);
```

```python
sns.jointplot(x="math score", y="reading score",  data=df, kind="hex", height=8, space=0.1, ratio=5);
```

```python
sns.jointplot(x="math score", y="reading score",  data=df, kind="kde", shade=True, hue="gender", alpha=0.3);
```

```python
sns.jointplot(x="math score", y="reading score",  data=df, kind="reg");
```

```python
sns.jointplot(x="math score", y="reading score",  data=df, kind="hist", bins=100, hue="gender");
```
